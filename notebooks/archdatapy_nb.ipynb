```python
from archdatapy import get_archdata, load_archdata
from pprint import pprint
import pandas as pd
```

```python
# Step 1: Download and get file paths
file_paths = get_archdata()
print("Available datasets and their file paths:")
pprint(file_paths)
```

```python
# Step 2: Load a specific dataset
dataset_name = 'Acheulean'  # Replace with the actual dataset name from file_paths keys
if dataset_name in file_paths:
    data = load_archdata(file_paths[dataset_name])
    pprint(data)  # Inspect the result from pyreadr
else:
    print(f"Dataset {dataset_name} not found.")
```
